```python
import sys
from PySide6.QtWidgets import QApplication, QWidget, QVBoxLayout, QLabel, QPushButton, QTextEdit
from PySide6.QtCore import Qt

class OldBoardPlaceholder(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Old RFID DAQ - placeholder")
        self.resize(700, 420)
        layout = QVBoxLayout(self)
        title = QLabel("Old RFID DAQ script placeholder")
        title.setAlignment(Qt.AlignCenter)
        title.setStyleSheet("font-size: 24px; font-weight: bold;")
        layout.addWidget(title)
        msg = QTextEdit()
        msg.setReadOnly(True)
        msg.setText(
            "This is a placeholder for the legacy / old RFID DAQ acquisition script.\n\n"
            "Replace this notebook in:\n"
            "DeepEcoHAB_dependencies/notebooks/old_board_placeholder.ipynb\n\n"
            "or edit main_menu_config.json to point to the final old-board notebook/script."
        )
        layout.addWidget(msg)
        btn = QPushButton("Close")
        btn.clicked.connect(self.close)
        layout.addWidget(btn)

def main():
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    w = OldBoardPlaceholder()
    w.show()
    return app.exec()

if __name__ == "__main__":
    raise SystemExit(main())
```
